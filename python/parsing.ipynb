```python
import pandas as pd
```

```python
data = pd.read_csv('./tb_product.csv')
```

```python
for i in range(len(data['p_category'])):
    temp = data['p_category'][i].replace('[', '').replace(']','').replace("'", '').split(',')
    data['category1'][i] = temp[0]
    data['category2'][i] = temp[1]
    if len(temp) > 2 :
        data['category3'][i] = temp[2]
    else :
        data['category3'][i] = '-'
```

```python
data.to_csv('tb.csv', sep=',', index = False)
```
